```python
import pickle
import matplotlib.pyplot as plt

with open('train_history31.pkl', 'rb') as f:
    history_bi = pickle.load(f)
with open('train_history32.pkl', 'rb') as f:
    history_uni = pickle.load(f)

plt.figure(figsize=(8, 5))
plt.plot(history_uni.history['loss'], label='Unidirectional')
plt.plot(history_bi.history['loss'], label='Bidirectional')
plt.title('Training Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_uni.history['val_loss'], label='Unidirectional')
plt.plot(history_bi.history['val_loss'], label='Bidirectional')
plt.title('Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history_uni.history['loss'], label='Train - Uni')
plt.plot(history_uni.history['val_loss'], label='Val - Uni', linestyle='--')
plt.plot(history_bi.history['loss'], label='Train - Bi')
plt.plot(history_bi.history['val_loss'], label='Val - Bi', linestyle='--')
plt.title('Training & Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
