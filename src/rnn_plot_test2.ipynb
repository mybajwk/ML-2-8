```python
import pickle
import matplotlib.pyplot as plt

with open('train_history21.pkl', 'rb') as f:
    history21 = pickle.load(f)
with open('train_history22.pkl', 'rb') as f:
    history22 = pickle.load(f)
with open('train_history23.pkl', 'rb') as f:
    history23 = pickle.load(f)

plt.figure(figsize=(8, 5))
plt.plot(history21.history['loss'], label='1 Layer - 64 Neuron')
plt.plot(history22.history['loss'], label='1 Layer - 128 Neuron')
plt.plot(history23.history['loss'], label='1 Layer - 256 Neuron')
plt.title('Training Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history21.history['val_loss'], label='1 Layer - 64 Neuron')
plt.plot(history22.history['val_loss'], label='1 Layer - 128 Neuron')
plt.plot(history23.history['val_loss'], label='1 Layer - 256 Neuron')
plt.title('Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history21.history['loss'], label='Train - 64 Neuron')
plt.plot(history21.history['val_loss'], label='Val - 64 Neuron', linestyle='--')
plt.plot(history22.history['loss'], label='Train - 128 Neuron')
plt.plot(history22.history['val_loss'], label='Val - 128 Neuron', linestyle='--')
plt.plot(history23.history['loss'], label='Train - 256 Neuron')
plt.plot(history23.history['val_loss'], label='Val - 256 Neuron', linestyle='--')
plt.title('Training & Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
