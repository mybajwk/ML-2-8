```python
import pickle
import matplotlib.pyplot as plt

with open('train_history11.pkl', 'rb') as f:
    history11 = pickle.load(f)
with open('train_history12.pkl', 'rb') as f:
    history12 = pickle.load(f)
with open('train_history13.pkl', 'rb') as f:
    history13 = pickle.load(f)

plt.figure(figsize=(8, 5))
plt.plot(history11.history['loss'], label='1 Layer - 64 Neuron')
plt.plot(history12.history['loss'], label='2 Layers - 64 Neuron')
plt.plot(history13.history['loss'], label='3 Layers - 64 Neuron')
plt.title('Training Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history11.history['val_loss'], label='1 Layer - 64 Neuron')
plt.plot(history12.history['val_loss'], label='2 Layers - 64 Neuron')
plt.plot(history13.history['val_loss'], label='3 Layers - 64 Neuron')
plt.title('Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(history11.history['loss'], label='Train - 1 Layer')
plt.plot(history11.history['val_loss'], label='Val - 1 Layer', linestyle='--')
plt.plot(history12.history['loss'], label='Train - 2 Layers')
plt.plot(history12.history['val_loss'], label='Val - 2 Layers', linestyle='--')
plt.plot(history13.history['loss'], label='Train - 3 Layers')
plt.plot(history13.history['val_loss'], label='Val - 3 Layers', linestyle='--')
plt.title('Training & Validation Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
